```python
import pandas as pd

df = pd.read_csv("/content/Export _  _ 3. August 2022 – 20. März 2023.csv", encoding="utf-16LE", low_memory=False)
df.head()
```

```python
df.columns
```

```python
languages = df.groupby("language").size()
display(languages)
```

```python
lang = {
"Deutsch":1087,
"Englisch":50,
"French":3,
}
lang=pd.Series(lang)
lang
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 10))
plt.title("User languages")
plt.pie(languages)
plt.legend(languages.index)
plt.show()
```
