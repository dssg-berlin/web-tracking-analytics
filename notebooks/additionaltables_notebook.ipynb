```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Search Engines

```python
df_searchengine = pd.read_csv("https://drive.google.com/uc?export=download&id=1zQMWnggV0nRDHGilWkHlbSxXoGc_r2oh", encoding="utf-16LE")
df_searchengine.head()
```

```python
plt.figure(figsize=(6,6))
plt.title("Search Engines Referred From")
plt.pie(df_searchengine["Besuche"])
plt.legend(df_searchengine["Bezeichnung"])
plt.show()
```
