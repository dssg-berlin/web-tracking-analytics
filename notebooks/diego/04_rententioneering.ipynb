```python
# Prepare columns for a dataset for retentioneering analysis
df_titles = df_main_table_unpivot[df_main_table_unpivot["variable"].isin(titles)]
df_timestamps = df_main_table_unpivot[
    df_main_table_unpivot["variable"].isin(timestamps)
]

# Add action number to the dataset to be used to join the resulting dataframes
df_titles["action_number"] = None
df_timestamps["action_number"] = None
for i, row in df_titles.iterrows():
    df_titles.at[i, "action_number"] = row["variable"].split(" ")[-1][:-1]

for i, row in df_timestamps.iterrows():
    df_timestamps.at[i, "action_number"] = row["variable"].split(" ")[-1][:-1]

# Rename columns to be used in retentioneering analysis
df_title = df_titles[["fingerprint", "value", "action_number"]]
df_title.columns = ["fingerprint", "event_full", "action_number"]
df_timestamp = df_timestamps[["fingerprint", "value", "action_number"]]
df_timestamp.columns = ["fingerprint", "timestamp", "action_number"]

# Convert timestamp to datetime
df_timestamp["timestamp"] = df_timestamp["timestamp"].map(
    lambda x: datetime.utcfromtimestamp(x)
)

# merge the two dataframes
df = pd.merge(df_title, df_timestamp, on=["fingerprint", "action_number"])
df.drop("action_number", axis=1, inplace=True)
```

```python
# Add event column for retentioneering analysis
df["event"] = None
for i, row in df.iterrows():
    if row["event_full"] == "/":
        # Replace homepage instead of /
        df.at[i, "event"] = "homepage"
    else:
        # Only last word of the title in event
        df.at[i, "event"] = row["event_full"].split("/")[-1].split(" ")[-1]
    # df.at[i, "event"] = row["event"].split("/")[-1].split(" ")[-1]
```

```python
# Save the resulting dataframe
df = df[["fingerprint", "event", "timestamp", "event_full"]]
df.to_csv("output/data_retentioneering.csv", index=False)
```
